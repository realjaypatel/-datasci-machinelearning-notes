```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','+'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time') #facit grid alternative you can still add information, hue = 'sex'
```
